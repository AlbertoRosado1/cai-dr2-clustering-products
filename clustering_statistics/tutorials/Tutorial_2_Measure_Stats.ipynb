## Running a Job Script 

If you need to generate a large amount measurements, it more convenient to submit as a job or run on an interactive node; the `desipipe_example.py` can do both. 
The `desipipe_example.py` contains some useful documentation within it, such as how to run the script, and what some important variables do and modify. Hopefully, the script is straightfoward to modify for your needs. Remember from Tutorial 1 that you can create your own `get_catalog_fn` function to manage filenames of clustering catalogs. 

Note: The `desipipe_example.py` script has different setups you can use, just uncomment or comment out unwanted arguments.

The first `mode` in which you can run the script is in `mode='interactive` by
```
salloc -N 1 -C "gpu&hbm80g" -t 04:00:00 --gpus 4 --qos interactive --account desi_g
source /global/common/software/desi/users/adematti/cosmodesi_environment.sh main
srun -n 4 python desipipe_example.py
```

The second `mode` is `mode='slurm'`, which uses `desipipe` to spawn and manage jobs. Below is the sequence of relevant `desipipe` commands
```
source /global/common/software/desi/users/adematti/cosmodesi_environment.sh main
python desipipe_example.py             # create the list of tasks
desipipe tasks -q example_job          # check the list of tasks
desipipe spawn -q example_job --spawn  # spawn the jobs
desipipe queues -q example_job         # check the queue
```
Note: On NERSC, after a certain amount of time jobs in `desipipe` no longer enter the queue. It might be a good idea to ocassionally do `squeue -u username` to double-check that jobs are still in queue or running.

https://github.com/cosmodesi/desipipe
