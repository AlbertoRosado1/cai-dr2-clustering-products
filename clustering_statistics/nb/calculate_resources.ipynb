# Full shape

## Load catalogs per region

```python
versions_dark   = ['holi-v3-altmtl']
versions_bright = ['holi-bgs-altmtl']

tracers_dark   = ['QSO', 'ELG_LOPnotqso', 'LRG']
tracers_bright = ['BGS_BRIGHT-21.35']

regions = ['NGC','SGC','N','NGCnoN','S','SGCnoDES','SnoDES','DES','ACT_DR6','PLANCK_PR4','GAL040','GAL060']

time_table_dark   = {version: {tracer: {region: 0 for region in regions} for tracer in tracers_dark} for version in versions_dark}
time_table_bright = {version: {tracer: {region: 0 for region in regions} for tracer in tracers_bright} for version in versions_bright}
    
profile_stats = np.loadtxt('../helper_scripts/profiling_full_shape_stats.txt', dtype=str)
for line in profile_stats:
    version = line[3]
    tracer  = line[1]
    region  = line[9]
    time    = line[11]
    # print(version)
    if 'bgs' in version:
        time_table_bright[version][tracer][region] = float(time)
    else:
        time_table_dark[version][tracer][region] = float(time)
        
# time_table_bright, time_table_dark
```

```python
# GOAL: Find the amount of GPU node hours needed to compute Pk + Bk for a desired number of mocks (`nmocks`)

versions = ['holi-v3-altmtl', 'holi-bgs-altmtl']
# regions = ['N','NGCnoN','S','SGCnoDES','SnoDES','DES','ACT_DR6','PLANCK_PR4','GAL040','GAL060']
regions = ['NGC','SGC','N','NGCnoN','S','SGCnoDES','SnoDES','DES','ACT_DR6','PLANCK_PR4','GAL040','GAL060']
nmocks = 1000 # number of mocks to process
total_time = 0
for version in versions:
    print(version)
    if 'bgs' in version:
        tracers = tracers_bright
    else:
        tracers = tracers_dark 
    for tracer in tracers:
        total_time_per_tracer = 0
        for region in regions:
            if 'bgs' in version:
                time = time_table_bright[version][tracer][region]
            else:
                time = time_table_dark[version][tracer][region]
            total_time += time
            total_time_per_tracer += time
            # print(f"{tracer} {region} {time*nmocks/60:2.1f} mins ({time*nmocks/60/60:2.1f} hours).")
        print(f"Need {total_time_per_tracer*nmocks/60:2.1f} mins ({total_time_per_tracer*nmocks/60/60:2.1f} hours) to compute {nmocks} Pk and Bk {tracer} for {regions}.")
print(f"Total {total_time*nmocks/60:2.1f} mins ({total_time*nmocks/60/60:2.1f} hours).")
            
```

```python
versions = ['holi-v3-altmtl']
regions = ['NGC','SGC']
tracers = ['LRG', 'ELGnotqso', 'QSO', 'LRGxQSO', 'LRGxELGnotqso', 'ELGnotqsoxQSO']

time_table  = {version: {tracer: {region: 0 for region in regions} for tracer in tracers} for version in versions}
    
profile_stats = np.loadtxt('../helper_scripts/profiling_local_png_stats.txt', dtype=str)
for line in profile_stats:
    version = line[3]
    tracer  = line[1]
    region  = line[8]
    time    = line[10]
    # print(version)
    time_table[version][tracer][region] = float(time)
# time_table
```

```python
# GOAL: Find the amount of GPU node hours needed to compute Pk for a desired number of mocks (`nmocks`)
nmocks = 1000 # number of mocks to process
total_time = 0
for version in versions:
    print(version)
    for tracer in tracers:
        total_time_per_tracer = 0
        for region in regions:
            time = time_table[version][tracer][region]
            total_time += time
            total_time_per_tracer += time
        print(f"Need {total_time_per_tracer*nmocks/60:2.1f} mins ({total_time_per_tracer*nmocks/60/60:2.1f} hours) to compute {nmocks} Pk {tracer} for {regions}.")
print(f"Total {total_time*nmocks/60:2.1f} mins ({total_time*nmocks/60/60:2.1f} hours).")
```
