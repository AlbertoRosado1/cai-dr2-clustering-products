```python
import sys

from pathlib import Path
import matplotlib.pyplot as plt

from desilike.samples import MCSamples, Profiles, plotting
from cosmoprimo.fiducial import AbacusSummit

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

params_toplot = ['omega_cdm', 'logA', 'h']

cosmo = AbacusSummit()
truth = {}
for p in params_toplot:
    try:
        truth[p] = cosmo[p]
    except:
        continue

def read_samples(fn, burnin=False, thin=False):
    if isinstance(fn, list):
        samples = MCSamples.concatenate([MCSamples.read(fn).remove_burnin(burnin if burnin else 0.) for fn in fn])
    else:
        samples = MCSamples.read(fn)
    if thin:
        samples = samples[::thin]
    return samples

plot_dir = Path('validation_abacus_lightcone')
plot_dir.mkdir(exist_ok=True)
ext = 'png'
```

## AbacusHF lightcone

```python
dirname = Path('/pscratch/sd/a/adematti/fits_lightcone_validation/abacus-hf-lc-dr2-v1.9/')
samples, legends = [], []
samples.append(read_samples([dirname / f'cosmo-base_QSO1-z0p8-2p1-ALL-S2-2f7ee649/samples_{i + 1:d}.h5' for i in range(4)], burnin=0.3, thin=10))
legends.append('$P(k)$ $0.8 < z < 2.1$')
samples.append(read_samples([dirname / f'cosmo-base_QSO-z0p8-1p6-ALL-S2-a9811541/samples_{i + 1:d}.h5' for i in range(4)], burnin=0.3, thin=10))
legends.append('$P(k)$ $0.8 < z < 1.6$')
samples.append(read_samples([dirname / f'cosmo-base_QSO-z1p6-2p1-ALL-S2-905f524d/samples_{i + 1:d}.h5' for i in range(4)], burnin=0.3, thin=10))
legends.append('$P(k)$ $1.6 < z < 2.1$')
plotting.plot_triangle(samples, params=params_toplot, title_limit=1, markers=truth, legend_labels=legends)
fig = plt.gcf()
fig.savefig(plot_dir / f'triangle.{ext}', bbox_inches='tight', pad_inches=0.1, dpi=200)
plt.show()
```
